## Importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Loading dataset

```python
df = pd.read_csv('top_10000_1960-now.csv')
df.head()
```

## EDA And Data Preprocessing

```python
df.info()
```

```python
# retaining the columns we gonna use
df = df[['Artist Name(s)','Track Name','Danceability','Energy','Key','Loudness','Mode','Speechiness','Acousticness','Instrumentalness','Liveness','Valence',
        'Tempo','Time Signature']]
```

```python
df.head()
```

```python
# dropping null values
df.dropna(inplace=True)
```

```python
# removing rows with similar records
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
columns = ['Danceability','Energy','Key','Loudness','Mode','Speechiness','Acousticness','Instrumentalness','Liveness',
           'Valence','Tempo','Time Signature']
```

```python
# convering columns into string then concatenating them into single column in order to build a content-based system
df['content'] = df['Danceability'].astype(str) + df['Energy'].astype(str) + df['Loudness'].astype(str) + df['Mode'].astype(str)+ df['Speechiness'].astype(str) + df['Acousticness'].astype(str) + df['Instrumentalness'].astype(str) + df['Liveness'].astype(str)+ df['Valence'].astype(str) + df['Tempo'].astype(str) + df['Time Signature'].astype(str)
```

```python
df.head()
```

```python
df.drop(columns,axis=1,inplace=True)
```

```python
df.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import sigmoid_kernel,cosine_similarity
```

```python
#creating a matrix
vect = TfidfVectorizer()
matrix = vect.fit_transform(df['content'])
```

```python
#defining similarity
sim = cosine_similarity(matrix)
```

## Performing recommendations

```python
#making recommendation
def reccomend(new_song):
    ind = df[df['Track Name'] == new_song].index[0]#Filters the DataFrame df to get the rows where the 'Track Name' is equal to new_song.
                                                   # Retrieves the index of the filtered DataFrame.
                                                   # Accesses the first element of this index.
    distance = sorted(list(enumerate(sim[ind])),reverse = True,key = lambda x: x[1])
    for i in distance[1:11]:
        print(f"{df['Track Name'].iloc[i[0]]} by {df['Artist Name(s)'].iloc[i[0]]}")
```

```python
reccomend("You Can't Always Get What You Want")
```
